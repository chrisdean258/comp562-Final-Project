```python
import pandas as pd
import numpy as np
from collections import defaultdict
from collections import Counter
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
df = pd.read_csv("equal.csv")
df["mode"] = 1 * (df["mode"] == "Major")
del df["artist_name"]
del df["track_id"]
del df["track_name"]
del df["key"]
del df["time_signature"]
df.head()
```

```python
for label in df:
    try:
        df[label] = df[label] / max(abs(a) for a in df[label])
    except TypeError:
        pass

df.head()
```

```python
def run_model(model, df, labels):
    trained = model.fit(df)
    counters = defaultdict(Counter)

    for predicted, actual in zip(trained.labels_, labels):
        counters[predicted][actual] += 1

    clust_labels = {}
    for cluster, counts in counters.items():
        clust_labels[cluster] = [a[0] for a in counts.most_common()]
    
    return trained, clust_labels 


def correct_top_n(labels, predicted, cluster_counts, top=1):
    correct = 0
    for label, prediction in zip(labels, predicted):
        if label in cluster_counts[prediction][:top]:
            correct += 1
    return correct / len(labels)


def tts_eval(model, df):
    train = df.sample(frac=0.9,random_state=200)
    test = df.drop(train.index)
    trained, cluster_counts = run_model(model, train.drop(["genre"], axis=1), train["genre"])
    predicted = trained.predict(test.drop(["genre"], axis=1))
    return correct_top_n(test["genre"], predicted, cluster_counts, top=3)
```

```python
# Values manually recorded in the cache to allow rerunning.
# For a clean run you need to delete them
cache = {
    25: 0.4549833461439918,
    125: 0.5398411478349987,
    625: 0.5830386881885729,
    3125: 0.6078401229823213,
    8125: 0.5933384575967204,
    15625: 0.5724827056110684,
    5625: 0.6017422495516269,
    4375: 0.6070202408403792,
    3750: 0.6077376377145786,
    3437: 0.6048168075839098,
    3281: 0.607122726108122,
    3203: 0.6099410709710479,
    3242: 0.6054317191903664,
    3222: 0.6022546758903408,
    3212: 0.6022034332564694,
    3207: 0.6108121957468614,
    3209: 0.6032795285677683,
    3208: 0.6072252113758647   
}

def cache_eval(num_clust):
    if num_clust in cache:
        return cache[num_clust]
    print(num_clust, end=" ")
    val = tts_eval(KMeans(n_clusters=num_clust), df)
    print(val)
    cache[num_clust] = val
    return val
        
num_clust = 25
best = 0
while True:
    val = cache_eval(num_clust)
    
    if val > best:
        num_clust *= 5
        best = val
    else:
        break
        
prev = 0
while True:
    c = list(sorted(cache.items()))
    m = max(cache.items(), key=lambda a: a[1])
    index = c.index(m)
    if c[index+1][1] > c[index-1][1]:
        low = c[index][0]
        high = c[index+1][0]
    else:
        low = c[index+1][0]
        high = c[index][0]
    mid = (low + high) // 2
    if mid == low or mid == high:
        print("Optimal at ", mid)
        break
    cache_eval(mid)
```
